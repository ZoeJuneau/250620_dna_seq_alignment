```python
import matplotlib.pyplot as plt
import numpy as np
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import seaborn as sns
from matplotlib.patches import Rectangle
import matplotlib.patches as mpatches

def create_beautiful_alignment_viz(seq1, seq2, seq1_name="Sequence 1", seq2_name="Sequence 2"):
    """
    Create a beautiful visualization of DNA sequence alignment
    """
    # Perform alignment
    alignments = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)
    alignment = alignments[0]
    
    aligned_seq1, aligned_seq2, score, begin, end = alignment
    
    # Calculate statistics
    matches = sum(1 for a, b in zip(aligned_seq1, aligned_seq2) if a == b and a != '-' and b != '-')
    mismatches = sum(1 for a, b in zip(aligned_seq1, aligned_seq2) if a != b and a != '-' and b != '-')
    gaps_seq1 = aligned_seq1.count('-')
    gaps_seq2 = aligned_seq2.count('-')
    identity = matches / (matches + mismatches) * 100 if (matches + mismatches) > 0 else 0
    
    # Create the visualization
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(16, 12))
    fig.suptitle('DNA Sequence Alignment Analysis', fontsize=16, fontweight='bold')
    
    # 1. Alignment Overview Bar
    alignment_length = len(aligned_seq1)
    chunk_size = max(1, alignment_length // 1000)  # Adjust for very long sequences
    
    # Create color-coded alignment bar
    colors = []
    for i in range(0, alignment_length, chunk_size):
        chunk1 = aligned_seq1[i:i+chunk_size]
        chunk2 = aligned_seq2[i:i+chunk_size]
        
        chunk_matches = sum(1 for a, b in zip(chunk1, chunk2) if a == b and a != '-')
        chunk_mismatches = sum(1 for a, b in zip(chunk1, chunk2) if a != b and a != '-' and b != '-')
        chunk_gaps = sum(1 for a, b in zip(chunk1, chunk2) if a == '-' or b == '-')
        
        if chunk_gaps > chunk_matches + chunk_mismatches:
            colors.append('#FF6B6B')  # Red for gaps
        elif chunk_matches > chunk_mismatches:
            colors.append('#4ECDC4')  # Teal for matches
        else:
            colors.append('#FFE66D')  # Yellow for mismatches
    
    # Plot alignment overview
    x_positions = np.arange(len(colors))
    ax1.bar(x_positions, [1]*len(colors), color=colors, width=1, edgecolor='none')
    ax1.set_xlim(0, len(colors))
    ax1.set_ylim(0, 1)
    ax1.set_title('Alignment Overview (Color-coded by Region Quality)', fontweight='bold')
    ax1.set_xlabel('Sequence Position (chunks)')
    ax1.set_yticks([])
    
    # Add legend
    match_patch = mpatches.Patch(color='#4ECDC4', label='Mostly Matches')
    mismatch_patch = mpatches.Patch(color='#FFE66D', label='Mostly Mismatches')
    gap_patch = mpatches.Patch(color='#FF6B6B', label='Gaps/Indels')
    ax1.legend(handles=[match_# patch, mismatch_patch, gap_patch], loc='upper right# ')
    
    # 2. Detailed alignment view (first 100#  bases)
    display_length = mi# n(100, alignment_length)
    
    # Creat# e matrix for heatmap
    matrix = np.zeros((2, display_length))# 
    base_to_num = {'A': 1, 'T': 2,#  'G': 3, 'C': 4, '-': 0}
    
    for i in range(display_# length):
        matrix[0, i] = base_to_num.get(aligned_seq1[i]# , 0)
        matrix[1# , i] = base_to_num.get(aligned_seq2[i], 0)
    
    # Custom colormap
    colors# _list = ['white', '#FF9999', '#99FF99', '#9999FF', '#FFFF99', '#FF# 99FF']
    cmap = plt.matplotlib.colors.ListedColormap(colors_list)
    
   #  im = ax2.imshow(matrix, cmap=cmap, aspect='auto', interpolation='nearest')
    ax2.set_title(f# 'Detailed Alignment View (First # {display_length} bases)', f# ontweight='bold')
    ax2.set_ylabel('Sequences# ')
    ax2.set_yticks([0, 1])
    # ax2.set_yticklabels([seq1_name, seq2_name])# 
    ax2.set_xlabel('Position# ')
    
    # Add base labels for shorter se# quences
    if display_length <= 50:
        ax2.set_xticks(range# (display_length))
        ax2.# set_xticklabels(range(1, display_length # + 1))
        
        # Add text annotations
        for i in range(display_length):
           #  ax2.text(i, 0, aligned_seq1[i], ha='center', va='center', fontsize=8, fontweight='bold')
            ax2.text(i, 1, aligned_seq2[i], ha='center', va='center', fontsize=8, fontweight='bold')
    
    # 3. Statistics Summary
    ax3.axis('off')
    
    # Create statistics text
    stats_text = f"""
    ALIGNMENT STATISTICS
    
    Alignment Score: {score:.0f}
    Sequence Length: {seq1_name}: {len(seq1)} bp, {seq2_name}: {len(seq2)} bp
    Aligned Length: {alignment_length} bp
    
    Matches: {matches} ({matches/alignment_length*100:.1f}%)
    Mismatches: {mismatches} ({mismatches/alignment_length*100:.1f}%)
    Gaps in {seq1_name}: {gaps_seq1} ({gaps_seq1/alignment_length*100:.1f}%)
    Gaps in {seq2_name}: {gaps_seq2} ({gaps_seq2/alignment_length*100:.1f}%)
    
    Sequence Identity: {identity:.1f}%
    """
    
    ax3.text(0.1, 0.9, stats_text, transform=ax3.transAxes, fontsize=12,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.8))
    
    # Add quality assessment
    if identity > 95:
        quality = "Excellent"
        color = "green"
    elif identity > 90:
        quality = "Very Good"
        color = "lightgreen"
    elif identity > 80:
        quality = "Good"
        color = "orange"
    else:
        quality = "Poor"
        color = "red"
    
    quality_text = f"Alignment Quality: {quality}"
    ax3.text(0.6, 0.7, quality_text, transform=ax3.transAxes, fontsize=14,
             fontweight='bold', color=color,
             bbox=dict(boxstyle="round,pad=0.3", facecolor=color, alpha=0.2))
    
    plt.tight_layout()
    plt.show()
    
    return aligned_seq1, aligned_seq2, score

def create_difference_plot(seq1, seq2, seq1_name="Sequence 1", seq2_name="Sequence 2"):
    """
    Create a plot highlighting differences between sequences
    """
    alignments = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)
    aligned_seq1, aligned_seq2, score, begin, end = alignments[0]
    
    # Find differences
    differences = []
    for i, (a, b) in enumerate(zip(aligned_seq1, aligned_seq2)):
        if a != b:
            differences.append((i, a, b))
    
    # Create plot
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Plot differences as scatter points
    positions = [d[0] for d in differences]
    y_seq1 = [0.6] * len(positions)
    y_seq2 = [0.4] * len(positions)
    
    # Color code by difference type
    colors = []
    for pos, a, b in differences:
        if a == '-' or b == '-':
            colors.append('red')  # Gaps
        else:
            colors.append('orange')  # Mismatches
    
    ax.scatter(positions, y_seq1, c=colors, s=100, alpha=0.7, label=f'{seq1_name} differences')
    ax.scatter(positions, y_seq2, c=colors, s=100, alpha=0.7, label=f'{seq2_name} differences')
    
    # Add connecting lines
    for pos in positions:
        ax.plot([pos, pos], [0.4, 0.6], 'k-', alpha=0.3)
    
    ax.set_ylim(0, 1)
    ax.set_xlim(0, len(aligned_seq1))
    ax.set_xlabel('Alignment Position')
    ax.set_title('Sequence Differences Visualization', fontweight='bold', fontsize=14)
    ax.set_yticks([0.4, 0.6])
    ax.set_yticklabels([seq2_name, seq1_name])
    
    # Add legend
    red_patch = mpatches.Patch(color='red', label='Gaps/Indels')
    orange_patch = mpatches.Patch(color='orange', label='Mismatches')
    ax.legend(handles=[red_patch, orange_patch])
    
    plt.tight_layout()
    plt.show()
    
    return differences

# Example usage:
# aligned_seq1, aligned_seq2, score = create_beautiful_alignment_viz(egfp_nt_seq, syfp2_nt_seq, 
#                                                                   "EGFP", "SYFP2")
# differences = create_difference_plot(egfp_nt_seq, syfp2_nt_seq, "EGFP", "SYFP2")
```
